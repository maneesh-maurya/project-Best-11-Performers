# Batting Summary

```python
#'''jason_file = pd.Series(pd.read_json('/content/drive/MyDrive/Business Analyst Projects/t20_wc_batting_summary.json')['battingSummary'].str[0]).to_dict()
#pd.json_normalize(jason_file)
#pd.DataFrame(jason_file).transpose()'''

#jason_file = pd.Series(pd.read_json('/content/drive/MyDrive/Business Analyst Projects/t20_wc_batting_summary.json')['battingSummary'].str[0]).to_dict().values()
#batting_Summary = pd.json_normalize(jason_file)
#batting_Summary
data = pd.read_json('/content/drive/MyDrive/Business Analyst Projects/t20_wc_batting_summary.json')
Team_detail = []
for i in data['battingSummary']:
  for j in i:
    Team_detail.append(j)

battingSummary1 = pd.DataFrame(Team_detail)
```

```python


battingSummary1.info()
# no missing info
```

```python
# making Dismissal colum in two category-  OUT / NOT OUT
#dismissal has no NULL value
def OUT_NOTOUT(x):
  if len(x) > 0:
    return 'Out'
  else:
    return 'Not-Out'

    

battingSummary1['dismissal'] = battingSummary1['dismissal'].apply(OUT_NOTOUT) #.apply(OUT_NOTOUT)[7]
battingSummary1
```

```python
# batsman name hace mny special charater and brckets like (c) which means they are captain, i do't want this
# lets remmove them

battingSummary2= battingSummary1['batsmanName'].str.replace('\(.+\)+|†+', '', regex = True) 
battingSummary2 
```

```python
battingSummary1['batsmanName'] = battingSummary2
# i need to merge both the table  batting and math 
# but the batting has no id ,
# we can take match id from match table then put in batting , then we can merge them
# but the problem is tha name of team can be is reverse order like team1 vs team2 , or team3 vs team 1
test ={}
for i in range(len(matchSummary['team1'])):
  key1 = matchSummary['team1'][i]+' '+'Vs'+' '+matchSummary['team2'][i]
  key2 = matchSummary['team2'][i]+' '+'Vs'+' '+matchSummary['team1'][i]
  
  test[key1] = matchSummary['match_ID'][i]
  test[key2] = matchSummary['match_ID'][i]
 

test

battingSummary1['match_ID'] = battingSummary1['match'].apply(lambda x: test[x])

battingSummary1
```

# Bowling Summary

```python
bowling_summary = pd.read_json('/content/drive/MyDrive/Business Analyst Projects/t20_wc_bowling_summary.json')
bowling_detail = []
for i in bowling_summary['bowlingSummary']:  # loadinf the json file
  for j in i:
    bowling_detail.append(j)

bowling_summary = pd.DataFrame(bowling_detail)
bowling_summary
# bowlers name looks fine so need to clean here
```

```python
# i need to merge both the table batting and math 
# but the bowling has no id ,
# we can take match id from match table then put in bowling , then we can merge them
# but the problem is tha name of team can be is reverse order like team1 vs team2 , or team3 vs team 1
bowl ={}
for i in range(len(matchSummary['team1'])):
  key1 = matchSummary['team1'][i]+' '+'Vs'+' '+matchSummary['team2'][i]
  key2 = matchSummary['team2'][i]+' '+'Vs'+' '+matchSummary['team1'][i]
  
  bowl[key1] = matchSummary['match_ID'][i]
  bowl[key2] = matchSummary['match_ID'][i]
 

bowl

bowling_summary['match_ID'] = bowling_summary['match'].apply(lambda x: bowl[x])

bowling_summary
```
